```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pl = 'viridis'
```

```python
def prod_errs(a, err = 0.1):
    x = []
    for i in a:
        x.append(i - err)
        x.append(i + err)
    return x
```

# p1 

```python
df = pd.DataFrame({
    'pora_dnia': ['rano'] * 4 + ['wieczorem'] * 4, 
    'pobudzenie': (['niskie'] * 2 + ['wysokie'] * 2) * 2,
    'wynik': prod_errs([3, 5, 5, 6])
    })
```

```python
df
```

```python
sns.catplot(x='pora_dnia', y='wynik', hue='pobudzenie', data=df, 
            kind='point', capsize=0.2,  markers=["o", "x"], linestyles=["-", "--"], palette=pl)

plt.savefig('pl1.png', dpi=150)
```
