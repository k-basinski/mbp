# Korelacje


```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import pearsonr
```

```python
d = {
    'x': np.random.normal(100, 15, 1000),
    'y1': np.random.normal(100, 15, 1000)
}

df = pd.DataFrame(d)

# correlated stuff
df['y2'] = (df.x * .4) + (df.y1 * .6)
df['y3'] = -(df.x * .2) + (df.y1 * .8)
df['y4'] = -(df.x * .5) + (df.y1 * .5)
```

```python
df
```

## v1


```python
sns.relplot(x='x', y='y1', data=df)
```

```python
r, p = pearsonr(df.x, df.y1)
print(f"r = {round(r, 2)}; p = {round(p, 3)}")
```

## v2


```python
sns.relplot(x='x', y='y2', data=df)
```

```python
r, p = pearsonr(df.x, df.y2)
print(f"r = {round(r, 2)}; p = {round(p, 3)}")
```

## v3


```python
sns.relplot(x='x', y='y3', data=df)
```

```python
r, p = pearsonr(df.x, df.y3)
print(f"r = {round(r, 2)}; p = {round(p, 3)}")
```
