# Korelacje

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
people = pd.read_csv("data/weight-height.csv")
less_arr = np.random.randint(0,10000, size = 300)
less_people = people.iloc[list(less_arr)]
```

## Waga i wzrost

```python
plt.scatter(less_people["Height"], less_people["Weight"], alpha=0.8)
plt.xlabel("waga")
plt.ylabel("wzrost")
plt.show()
```

## Waga i wzrost - z płcią

```python
# wymień płeć na 0-1
l_ppl = less_people.replace({'Gender': {'Male' : 0, 'Female' : 1}})
```

```python
plt.scatter(l_ppl["Height"], l_ppl["Weight"], 
            alpha=0.8, 
            c = l_ppl["Gender"],
            cmap = "viridis")
plt.xlabel("waga")
plt.ylabel("wzrost")
plt.show()
```
